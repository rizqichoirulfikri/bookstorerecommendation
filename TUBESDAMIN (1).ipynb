# Preprocessing

# **Merge Dataset**

```python
# Merge Dataset
df = df_items.merge(df_transactions, left_on='itemID', right_on='itemID')
```

# **Missing Values**

```python
df.isnull()
```

```python
# Mengisi nilai Null
df.fillna('', inplace=True)
df.isnull().sum()
```

```python
df.info()
```

```python
#menghitung click basket and order for each books
total_trx = df.groupby('itemID').agg({'click' : 'sum'}).rename(columns={'click' : 'total interactions'})
# df_apply = 
```

```python
df = df.merge(total_trx, on='itemID')
df.head()
```

**Collaborative Filtering**

Customer Item

```python
def make_customer_item_matrix(data):
    # customer_item_matrix
    customer_item_matrix = data.pivot_table(
        index='title', 
        columns='itemID', 
        values='click', 
        aggfunc='sum'
    )
    customer_item_matrix = customer_item_matrix.applymap(lambda x: 1 if x > 0 else 0) # klik buku maka session nilai nya 1 dan sebaliknya
    return customer_item_matrix
```

```python
df_test = df[:10000]
```

```python
customer_item_matrix = make_customer_item_matrix(df_test)
```

```python
customer_item_matrix 
```

**Item to Item Similarity Matrix**

```python
# Ngitung similarity (kemiripan) antar item
item_matrix=pd.DataFrame(cosine_similarity(customer_item_matrix))
```

```python
# 1.0 (matrix identitas)
item_matrix
```

```python
# penamaan baris dan kolom
item_matrix.columns = customer_item_matrix.index #kolom yang diambil dari index customer item matrix
item_matrix['title'] = customer_item_matrix.index #baris yang diambil dari index customer item matrix
item_matrix = item_matrix.set_index('title') #di set indexnya
```

```python
item_matrix
```

```python
top5_item = list(item_matrix.loc['Ace of Shades'].sort_values(ascending=False).iloc[:5].index)
top5_item
```

```python

df_test.loc[df_test['title'].isin(top5_item), ['itemID', 'title',		'publisher',	'main topic']
].drop_duplicates().set_index('title').loc[top5_item]
```
